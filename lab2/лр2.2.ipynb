```python
import pandas as pd
import pandasql as ps


data = pd.read_csv('user_device.csv', sep=",")
data1 = pd.read_csv('user_usage.csv', sep=",")
data.head()
```

```python
data1.head()
```

```python
result = pd.merge(data1,
                 data[['use_id', 'platform', 'device']],
                 on='use_id')
result.head()
```

```python
%timeit result.head()
```

```python
join_query="""SELECT w.outgoing_mins_per_month, w.outgoing_sms_per_month, w.monthly_mb, t.use_id, t.platform, t.device
FROM data1 AS w JOIN data AS t
ON w.use_id=t.use_id"""
result1 = ps.sqldf(join_query)
result1.head()
```
